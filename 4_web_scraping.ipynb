```python
final = pd.DataFrame()

for j in range(1, 500):
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers=headers).text

    soup = BeautifulSoup(webpage, 'lxml')
    company = soup.find_all(class_='companyCardWrapper')

    name = []
    sector = []
    location = []
    rating = []
    reviews = []
    jobs = []

    
    for i in company:
        try: 
            name.append(i.find_all('h2')[0].text.strip())
        except:
            name.append(np.nan)

        try:
            sector.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip().split('|')[0])
        except:
            sector.append(np.nan)
      
        try:
            location.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip().split('|')[1].strip())
        except:
            location.append(np.nan)

        try: 
            rating.append(i.find_all(class_='rating_text')[0].text.strip())
        except:
            rating.append(np.nan)

        try: 
            reviews.append(i.find('span', class_='companyCardWrapper__ActionCount').text)
        except:
            reviews.append(np.nan)

        try: 
            jobs.append(i.find_all('span', class_='companyCardWrapper__ActionCount')[3].text)
        except:
            jobs.append(np.nan)

    df = pd.DataFrame({
        'name':name,
        'sector': sector,
        'location': location,
        'rating': rating,
        'reviews': reviews,
        'jobs': jobs,
    })
    
    final = pd.concat([final, df], ignore_index=True)
```
