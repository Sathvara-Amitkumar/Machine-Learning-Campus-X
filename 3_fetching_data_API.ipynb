```python
for i in range(1, 500):
    res = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=9c52ef5a134d2273347f9db339317aae&language=en-US&page={}&region=AT".format(i))
    temp_df = pd.DataFrame(res.json()['results'])[['id','title','overview','release_date','vote_average']]
    df = pd.concat([df, temp_df], ignore_index=True)
```
